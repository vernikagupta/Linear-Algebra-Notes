#### Symmetric matrix is a matrix which is equal to its tramspose.
A = A.T
1. symmetric matrix when applied on a vector produce scale + shear effect.
2. Symmetric matrix produce real eignvectors which are orthogonal to each other
3. Vectors that lie in the direction of eignvector get scaled. no rotation, no shear

```python
import matplotlib.pyplot as plt
import numpy as np

# Symmetric matrix
sym_matrix = np.array([[2, 1], [1, 2]])

# Eigen decomposition
eigvals, eigvecs = np.linalg.eig(sym_matrix)

# Set up plot
fig, ax = plt.subplots(figsize=(5, 5))
ax.set_aspect('equal')
ax.axhline(0, color='black', linewidth=0.5)
ax.axvline(0, color='black', linewidth=0.5)

# Example vectors from different quadrants
vectors = [np.array([1, 0]), np.array([0, 1]), np.array([1, 1]), np.array([-1, 1])]

# Plot original vectors (blue)
for v in vectors:
    ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='blue', alpha=0.5)

# Plot transformed vectors (red)
for v in vectors:
    tv = sym_matrix @ v
    ax.quiver(0, 0, tv[0], tv[1], angles='xy', scale_units='xy', scale=1, color='red')

# Plot eigenvectors (green)
for i in range(len(eigvals)):
    eigvec = eigvecs[:, i] * eigvals[i]
    ax.quiver(0, 0, eigvec[0], eigvec[1], angles='xy', scale_units='xy', scale=1, color='green', width=0.005)

ax.set_xlim(-4, 4)
ax.set_ylim(-4, 4)
ax.set_title("Symmetric Matrix Transformation\nBlue: original | Red: transformed | Green: eigenvectors")
plt.grid(True)
plt.show()
```

Let's say matrix A = [1 2 ; 2 1] and Vector is [2 3] as a column vector


```python
import matplotlib.pyplot as plt
import numpy as np

# Matrix and vector
A = np.array([[1, 2], [2, 1]])
v = np.array([2, 3])

# Transform vector
tv = A @ v

# Prepare plot
fig, ax = plt.subplots(figsize=(5, 5))
ax.set_aspect('equal')
ax.axhline(0, color='black', linewidth=0.5)
ax.axvline(0, color='black', linewidth=0.5)

# Plot original vector
ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Original [2, 3]')

# Plot transformed vector
ax.quiver(0, 0, tv[0], tv[1], angles='xy', scale_units='xy', scale=1, color='red', label='Transformed [8, 7]')

# Other vectors to show shear effect
vectors = [np.array([1, 0]), np.array([0, 1]), np.array([1, 1]), np.array([-1, 1])]
for vec in vectors:
    tvec = A @ vec
    ax.quiver(0, 0, vec[0], vec[1], angles='xy', scale_units='xy', scale=1, color='gray', alpha=0.5)
    ax.quiver(0, 0, tvec[0], tvec[1], angles='xy', scale_units='xy', scale=1, color='orange', alpha=0.5)

ax.set_xlim(-10, 10)
ax.set_ylim(-10, 10)
ax.legend()
ax.set_title("Symmetric Matrix [[1, 2], [2, 1]]\nBlue=Original | Red=Transformed | Gray/Orange=Other examples")
plt.grid(True)
plt.show()
```

#### When we are using shear matrix where only one off diagnol no is non zero, then it produce only pure shear.
1. It means length of one coordinate axis stays same and only coordiante axis get tilted, (slid parallel), but area is preserved.
2. Matrix form (example for x‑shear): [ 1 s; 0 1]
3. First column:
[1,0] → x-axis unchanged in length.
4. Second column 
[s,1] → y-axis tilted.

Important:

Pure shear does not mean “no vector changes length.”

Most vectors will change length unless they are aligned with one of the coordinate axes.

The scaling effect depends on the direction of the vector.

Why it’s “pure”:

No uniform scaling applied to the whole space.

Determinant = 1 → no change in total area.



```python
import matplotlib.pyplot as plt
import numpy as np

# Matrix and vector
A = np.array([[1, 2], [0, 1]])
v = np.array([2, 3])

# Transform vector
tv = A @ v

# Prepare plot
fig, ax = plt.subplots(figsize=(5, 5))
ax.set_aspect('equal')
ax.axhline(0, color='black', linewidth=0.5)
ax.axvline(0, color='black', linewidth=0.5)

# Plot original vector
ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Original [2, 3]')

# Plot transformed vector
ax.quiver(0, 0, tv[0], tv[1], angles='xy', scale_units='xy', scale=1, color='red', label='Transformed [8, 3]')

# Other vectors to show shear effect
vectors = [np.array([1, 0]), np.array([0, 1]), np.array([1, 1]), np.array([-1, 1])]
for vec in vectors:
    tvec = A @ vec
    ax.quiver(0, 0, vec[0], vec[1], angles='xy', scale_units='xy', scale=1, color='gray', alpha=0.5)
    ax.quiver(0, 0, tvec[0], tvec[1], angles='xy', scale_units='xy', scale=1, color='orange', alpha=0.5)

ax.set_xlim(-10, 10)
ax.set_ylim(-10, 10)
ax.legend()
ax.set_title("Shear Matrix [[1, 2], [0, 1]]\nBlue=Original | Red=Transformed | Gray/Orange=Other examples")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Matrix
A = np.array([[1, 2],
              [2, 1]])

# Original vector
v = np.array([2, 3])

# Transform the vector
tv = A @ v

# Eigen decomposition
eigvals, eigvecs = np.linalg.eig(A)

print("Eigenvalues:", eigvals)
print("Eigenvectors:\n", eigvecs)

# Plot
fig, ax = plt.subplots(figsize=(6, 6))
ax.set_aspect('equal')
ax.axhline(0, color='black', linewidth=0.5)
ax.axvline(0, color='black', linewidth=0.5)

# Original vector
ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Original [2, 3]')

# Transformed vector
ax.quiver(0, 0, tv[0], tv[1], angles='xy', scale_units='xy', scale=1, color='red', label='Transformed [8, 7]')

# Eigenvectors
for i in range(len(eigvals)):
    ev = eigvecs[:, i] * eigvals[i]  # scale by eigenvalue for visibility
    ax.quiver(0, 0, ev[0], ev[1], angles='xy', scale_units='xy', scale=1, color='green', width=0.005, label=f'Eigenvector {i+1}')

ax.set_xlim(-10, 10)
ax.set_ylim(-10, 10)
ax.legend()
ax.set_title("Eigenvectors and Transformation Effect")
plt.grid(True)
plt.show()
```

If your vector lies exactly along one of the green eigenvector directions of the matrix:

It will not shear or change direction — it will stay on the same line through the origin.

The only change will be in its length (scaling), and possibly its sign (flip) if the eigenvalue is negative.
